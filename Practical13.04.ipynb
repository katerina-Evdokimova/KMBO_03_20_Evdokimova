```julia
function rashirennoe(a) # 12
    count = 0
    
    for i in 1:length(a)
        if a(i,length(a)-1) == 0 & a(i,length(a))!=0
            return nothing
        else
            count += 1
        end
    end
    
    #функция из 6 задачи, где она преобразовывает в стуаенчатую обратно
    
    x = collect(length(a))
    for k in 1:length(a)
        if k<=count 
            for t in count+1:length(a)
                x[k] += a[k][t]
            end
        else
            x[k] = 1
        end
    end
    
    return x
end
```
