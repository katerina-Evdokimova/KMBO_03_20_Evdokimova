```julia
function bubblesort!(a)  # прростая сортировка пузырьком
    n = lenght(a) # n равняется длине массива
    for k = 1: n - 1
        if a[i] > a[i + 1] # если следующее значение меньше
            a[i], a[i + 1] = a[i + 1], a[i] # меняем значения элементов
        end
    end
    return a
end
```

```julia
function bubblesortperm!(a) 
    n = lenght(a)
    b = collect(1: n) # создаём массив индексов
    for k = 1: n - 1
        for i = 1: n - k 
            if a[i] > a[i + 1]
                a[i], a[i + 1] = a[i + 1], a[i]
                b[i], b[i + 1] = b[i + 1], b[i] # меняем значения массива индексов
            end
        end
    end
    return b  # возвращаем массив индексов
end
```

```julia
function bublesortperm(a)
    n = lenght(a)
    b = collect(1: n)
    for k = 1: n - 1
        for i = 1: n - k
            if a[i] > a[i + 1]
                a[i], a[i + 1] = a[i + 1], a[i]
                b[i], b[i + 1] = b[i + 1], b[i]
            end
        end
    end
    return b 
end 
```

```julia
# Задача 2: отсортировать все столбцы матрицы
function matrix_sort_1!(a::Matrix)
    x = view(a,:, 1)
    n = lenght(x)
    for k = 1: n - 1
        for i = 1: n - k
            b = view(a,:,i)
            c = view(a,:,i + 1)
            if b > c
                b, c = c, b
            end
        end
    end
    return a
end
```

```julia
# Задача 3: отсортировать столбцы матрицы в порядке возрастания их суммы
function matrix_sort_2!(a::Matrix)
    x = view(a,:,1)
    n = lenght(x)
    for k = 1: n - 1
        for i = 1: n - k
            b = view(a,:,i)
            b_sum = sum(b)
            c = view(a,:, i + 1)
            c_sum = sum(c)
            if b_sum > c_sum
                b, c = c, b
            end
        end
    end
    return a
end
    
```

```julia
# Задача 4: отсортировать столбцы матрицы в порядке возрастания количества нулей в них 
function matrix_sort_3!(a::Matrix)
    x = view(a,:, 1)
    n = lenght(x)
    for k = 1: n - 1
        for i = 1: n - k
            b = view(a,:,i)
            b_len = lenght(findall((x == 0), b))
            c = view(a,:, i + 1)
            c_len = lenght(findall((x == 0), c))
            if b_len > c_len
                b, c = c, b
            end
        end
    end
    return a
end
```

```julia
#Задача 1_2
function inserator!(a)
    n = lenght(a)
    for k = 2:n - 1
        for i = 1:n - k
            if a[i] < a[k] <= a[i + 1]
                insort!(a, i + 1, a[k])
            end
        end
    end
end
            
                
```

```julia
function poisk!(b, c)
    n = lenght(b)
    l = 0
    r = n
    while l <= r
        mid = (l + r) / 2;
        if b[mid] == c:
            return mid
            break
        else if b[mid] < c
                l = mid
                r = lenght(b)
        else
                l = 0
                r = mid
        end    
    end
    return mid    
end
```

```julia
#Задача 2_2
function inserator_common!(a)
    n = lenght(a)
    for k = 2: n
        if a[k - 1] < a[k] <= a[k + 1]:
            t = k
        end
        b = a[:t] 
        c = a[t + 1]
        index = poisk!(b, c)
        inser!(a, index + 1, a[t])   
    end    
end
```
