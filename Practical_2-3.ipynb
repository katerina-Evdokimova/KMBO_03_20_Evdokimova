```julia
# Задача 1 :
function insertsort!(A)
    n = length(A)

    for k = 2:n
        for i = 1:n-k
            if A[i]< A[k] <= A[i+1]
                insert!(A, i + 1,A[k])
            end
        end
    end
end
```

```julia
# Задача 2 :
function insertsort_poisk!(A)
    n = length(A)
    for k = 2:n
        if A[k-1]<A[k]<=A[k+1]
            t = k
        end
        index = poisk!(A[:t], A[t+1])
        insert!(A, index + 1 , A[t])
    end
    
end

function poisk!(b, c)
    n = length(b)
    l = 0
    while l<=n
        mid = (l+n)/2
        if b[mid] == c
            return mid
            break 
        elseif b[mid] < c
            l = mid
            n = length(b)
        else 
            l = 0
            n = mid
        end
    end
    return mid
end
```
